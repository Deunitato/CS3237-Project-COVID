```python
import json
from collections import defaultdict
import matplotlib.pyplot as plt
from os import system, name 
```

##please change the variable below##

```python
filename = 'demo/stab_05112020.json'
moveLabel = 2
```

```python

dataSet = defaultdict(list)
count = 0
with open(filename, 'r') as f:
    for line in f:
        res = json.loads(line)
        
        dataSet["accelX"].append(float(res["accelX"]))
        dataSet["accelY"].append(float(res["accelY"]))
        dataSet["accelZ"].append(float(res["accelZ"]))
        dataSet["gyroX"].append(float(res["gyroX"]))
        dataSet["gyroY"].append(float(res["gyroY"]))
        dataSet["gyroZ"].append(float(res["gyroZ"]))
        dataSet["time"].append(count)
        count +=1
```

```python
def clear(): 
  
    # for windows 
    if name == 'nt': 
        _ = system('cls') 
  
    # for mac and linux(here, os.name is 'posix') 
    else: 
        _ = system('clear') 
  
```

```python
def plot(dataSet,timeStart,timeEnd):
    plt.plot(dataSet["time"][timeStart:timeEnd],dataSet["accelX"][timeStart:timeEnd])

    plt.title('A tale of 3 subplots')
    plt.ylabel('X acceleration')
    plt.show()

    plt.plot(dataSet["time"][timeStart:timeEnd],dataSet["accelY"][timeStart:timeEnd])
    plt.xlabel('time (s)')
    plt.ylabel('Y acceleration')
    plt.show()

    plt.plot(dataSet["time"][timeStart:timeEnd],dataSet["accelZ"][timeStart:timeEnd])
    plt.xlabel('time (s)')
    plt.ylabel('Z acceleration')

    plt.show()

    plt.plot(dataSet["time"][timeStart:timeEnd],dataSet["gyroX"][timeStart:timeEnd])

    plt.title('A tale of 3 subplots')
    plt.ylabel('X Gryo')
    plt.show()

    plt.plot(dataSet["time"][timeStart:timeEnd],dataSet["gyroY"][timeStart:timeEnd])
    plt.xlabel('time (s)')
    plt.ylabel('Y Gryo')
    plt.show()

    plt.plot(dataSet["time"][timeStart:timeEnd],dataSet["gyroZ"][timeStart:timeEnd])
    plt.xlabel('time (s)')
    plt.ylabel('Z Gryo')

    plt.show()
```

```python
starttime,startMove,EndMove = 0,0,0
while(starttime <=count):
    clear()
    plot(dataSet,starttime,starttime+70)
    startMove,EndMove = map(int,input("Enter the startmove and endmove : ").split())
    newStartTime = int(input("Enter the startTime : ") or starttime)
    print(startMove,EndMove ,newStartTime)
    for i in range(starttime,newStartTime if newStartTime <count else count):
        if(i in range(startMove,EndMove)):
            dataSet["label"].append(moveLabel)
        else:
            dataSet["label"].append(0)
    starttime = newStartTime
```

```python
with open('label/stab.txt', 'a+') as filehandle:
    for listitem in dataSet["label"]:
        filehandle.write('%s\n' % listitem)
```
